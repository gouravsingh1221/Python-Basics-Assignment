Q.6  Why are lists mutable, but tuples are immutable?

Q.7  What is the difference between “==” and “is” operators in Python?
 - The == operator checks if two values are the same. It looks at what is inside the variables. For example, if two lists have the same items, == will return True. On the other hand, the is operator checks if two variables point to the same object in memory. Even if two lists have the same values, they are different objects, so is will return False. But if one variable is just another name for the same object, then is will return True. So, == checks for same value, and is checks for same identity .

Q.8  What are logical operators in Python?

 - Logical operators in Python are used to make decisions based on conditions. They help combine two or more conditions and return True or False.

Q.9  What is type casting in Python?
 - In Python, type casting means changing one data type to another.


Q.10  What is the difference between implicit and explicit type casting?
 - Implicit type casting happens automatically when Python changes the type for you. On the other hand, explicit type casting is when you do it yourself by using functions like int(), float(), or str(). This is done when you want to be sure about the type change. So, the main difference is that implicit casting is done by Python, while explicit casting is done by the programmer.

Q.11  What is the purpose of conditional statements in Python?

- The purpose of conditional statements in Python is to help the program make decisions. They allow the program to choose what to do based on certain conditions.

Q.12  How does the elif statement work?
- The elif statement in Python means "else if", and it is used to check another condition if the first if condition is not true. It helps you check many different options one by one. Python will check the if condition first. If that is false, it checks the elif. If the elif is true, it runs that part of the code. You can also add an else at the end if none of the conditions are true.

Q.13 8 What is the difference between for and while loops?
- he for loop and while loop are both used to repeat code in Python, but they work in different ways. A while loop is used when you don’t know exactly how many times you need to repeat, but you want to keep going as long as a condition is true.A for loop is used when you know how many times you want to repeat something. It goes through a list, range, or any sequence.

Q.14 8 Describe a scenario where a while loop is more suitable than a for loop.

- A while loop is more suitable when you don’t know how many times the loop should run and you want to keep going until a certain condition is met. For example, if you are asking a user to enter a password until they get it right, you don’t know how many tries they will need. In this case, a while loop is better because it will keep asking the user again and again until the correct password is entered. So, a while loop is useful when the number of repetitions is not fixed.

```python
# Q1 Write a Python program to print "Hello, World!
print("Hello, World!")

# Q2 Write a Python program that displays your name and age

name = "Gourav singh baloria"
age = 20

print("My name is", name)
print("I am", age , "years old")

# Q3 Write code to print all the pre-defined keywords in Python using the keyword library
import keyword

print("Python keywords are:")
print(keyword.kwlist)

# Q4  Write a program that checks if a given word is a Python keyword.

word = input("Enter a word: ")

if keyword.iskeyword(word):
    print(f"'{word}' is a Python keyword.")
else:
    print(f"'{word}' is NOT a Python keyword.")

# Q5 Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.
# List example (mutable)
my_list = [1, 2, 3]
print("Original list:", my_list)
my_list[0] = 10
print("Modified list:", my_list)

# Tuple example (immutable)
my_tuple = (1, 2, 3)
print("\nOriginal tuple:", my_tuple)

try:
    my_tuple[0] = 10  # This will cause an error
except TypeError:
    print("You cannot change a tuple. It's immutable.")

print("Program continues running even after tuple error.")

# Q.6 Write a function to demonstrate the behavior of mutable and immutable arguments.
def change_things(my_list, my_number):
    my_list.append(10)      # trying to change the list
    my_number = my_number + 5  # trying to change the number
    print("Inside function:")
    print("List:", my_list)
    print("Number:", my_number)

# original values
numbers = [1, 2, 3]
age = 20

change_things(numbers, age)

print("\nOutside function:")
print("List:", numbers)   # list is changed
print("Number:", age)     # number is still the same

# Q.7  Write a program that performs basic arithmetic operations on two user-input numbers.
# Get input from user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform operations
print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)

# Check for division by zero
if num2 != 0:
    print("Division:", num1 / num2)
else:
    print("Division: Cannot divide by zero")

# Q.8 Write a program to demonstrate the use of logical operators.
# Get user input
age = int(input("Enter your age: "))
has_id = input("Do you have an ID? (yes/no): ")

# Use logical operators

# 'and' operator
if age >= 18 and has_id == "yes":
    print("You are allowed to enter.")
else:
    print("You are under age or have no ID, not allowed.")

# 'or' operator
if age >= 18 or has_id == "yes":
    print("You meet at least one condition.")

# 'not' operator
if not (has_id == "yes"):
    print("You must bring your ID.")

# Q.9 Write a Python program to convert user input from string to integer, float, and boolean types.
# Get input from user
user_input = input("Enter a value: ")

# Convert to integer
try:
    int_value = int(user_input)
    print("Integer:", int_value)
except ValueError:
    print("Cannot convert to integer.")

# Convert to float
try:
    float_value = float(user_input)
    print("Float:", float_value)
except ValueError:
    print("Cannot convert to float.")

# Convert to boolean
bool_value = bool(user_input)
print("Boolean:", bool_value)

# Q.10 . Write code to demonstrate type casting with list elements.
# List of strings (like user input)
string_list = ["1", "2", "3", "4"]

# Convert each string to an integer
int_list = [int(item) for item in string_list]

print("Original list (strings):", string_list)
print("After type casting to integers:", int_list)

# Q.11  Write a program that checks if a number is positive, negative, or zero.
# Get number from user
num = float(input("Enter a number: "))

# Check the number
if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")

# Q.12  Write a for loop to print numbers from 1 to 10.
for i in range(1, 11):
    print(i)

# Q.13  Write a Python program to find the sum of all even numbers between 1 and 50.
total = 0

for i in range(1, 51):
    if i % 2 == 0:
        total += i

print("Sum of even numbers from 1 to 50 is:", total)

#Q.14  Write a program to reverse a string using a while loop.
# Get input from user
text = input("Enter a string: ")

# Initialize variables
reversed_text = ""
index = len(text) - 1

# Use while loop to reverse the string
while index >= 0:
    reversed_text += text[index]
    index -= 1

# Print the result
print("Reversed string:", reversed_text)

#Q15 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
# Get number from user
num = int(input("Enter a number: "))
factorial = 1
i = 1

# Check if the number is negative
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    while i <= num:
        factorial *= i
        i += 1
    print("Factorial of", num, "is:", factorial)
# COMPLETED

```
